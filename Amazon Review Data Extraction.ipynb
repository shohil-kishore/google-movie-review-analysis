```python
# From the Amazon Data website - parse data into a valid DF.
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

meta_df = getDF('/Users/shohilkishore/Desktop/Work/Gabrielle Peko/Amazon Data/meta_Movies_and_TV.json.gz')
```
